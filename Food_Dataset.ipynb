```python
prep_time_threshold = df['prep_time'].quantile(0.95)
cook_time_threshold = df['cook_time'].quantile(0.95)

df['prep_time'] = df['prep_time'].clip(upper=prep_time_threshold)
df['cook_time'] = df['cook_time'].clip(upper=cook_time_threshold)
```

```python
df = df[df['prep_time'] <= prep_time_threshold]
df = df[df['cook_time'] <= cook_time_threshold]
```

```python
df['prep_time'].fillna(df['prep_time'].median(), inplace=True)
df['cook_time'].fillna(df['cook_time'].median(), inplace=True)
```

<h3>Average Preparation and Cooking Time</h3>

```python
Average_prep_time = df.groupby('diet')[['prep_time','cook_time']].mean().reset_index()
print(Average_prep_time)
```

```python
x_point = Average_prep_time['diet']
y1_point = Average_prep_time['prep_time']
y2_point = Average_prep_time['cook_time']

plt.plot(x_point ,y1_point ,y2_point ,marker = 'o')
plt.xlabel('Diet')
plt.ylabel('Average Preparation Time')
plt.title('Average Preparation & cooking Time By Diet')
plt.legend(['Preparation Time','Cooking Time'])
plt.show()
```

Non-vegetarian dishes take more time overall (both preparation and cooking) compared to vegetarian dishes.

<h3>Dite Prefered by each Region</h3>

```python
No_of_dishes = df.groupby(['region','diet'])['name'].count().reset_index()
pivot_df = No_of_dishes.pivot(index='region', columns='diet', values='name').fillna(0)
print(pivot_df)
```

```python
# Extract data for plotting
regions = pivot_df.index
vegetarian_counts = pivot_df['vegetarian']
non_vegetarian_counts = pivot_df['non vegetarian']

# Define bar width and positions
bar_height = 0.36
r1 = np.arange(len(regions))
r2 = [x + bar_height for x in r1]

# Plot clustered bar chart
plt.figure(figsize=(12, 6))
plt.barh(r1, vegetarian_counts, color='g',height =bar_height,  edgecolor='grey', label='Vegetarian')
plt.barh(r2, non_vegetarian_counts, color='r',height =bar_height, edgecolor='grey', label='Non-Vegetarian')
plt.yticks([r + bar_height / 2 for r in range(len(regions))], regions)

plt.xlabel('Count of Dishes')
plt.ylabel('Region')
plt.title('Total number of dishes from each Region')
plt.show()
```

The West region stands out as having the highest variety or number of recorded dishes, with a significant majority being vegetarian.

Non-vegetarian dishes are relatively low in count across all regions but are somewhat more noticeable in North East and North regions compared to others.
